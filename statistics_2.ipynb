<b>–ó–∞–¥–∞–Ω–∏–µ 1.</b> –ò–∑ –ø–∞—Ä—Ç–∏–∏ –±–∞—Ç–∞—Ä–µ–µ–∫ —Å –Ω–æ–º–∏–Ω–∞–ª—å–Ω—ã–º –Ω–∞–ø—Ä—è–∂–µ–Ω–∏–µ–º 12 –í –Ω–∞—É–≥–∞–¥ –≤—ã–±—Ä–∞–Ω–æ —Å–µ–º—å —à—Ç—É–∫. –°–æ–≥–ª–∞—Å–Ω–æ –ø–æ–∫–∞–∑–∞–Ω–∏—è–º –≤—ã—Å–æ–∫–æ—Ç–æ—á–Ω–æ–≥–æ –≤–æ–ª—å—Ç–º–µ—Ç—Ä–∞, —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–∏ –≤—ã–¥–∞–≤–∞–µ–º–æ–µ –∏–º–∏ –Ω–∞–ø—Ä—è–∂–µ–Ω–∏–µ —Å–æ—Å—Ç–∞–≤–∏–ª–æ: 12,9; 11,6; 13,5; 13,9; 12,1; 11,9; 13,0 –í —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ. –ú–æ–∂–Ω–æ –ª–∏ —Å—á–∏—Ç–∞—Ç—å, —á—Ç–æ —Å—Ä–µ–¥–Ω–µ–µ –Ω–∞–ø—Ä—è–∂–µ–Ω–∏–µ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ —Ä–∞–≤–Ω–æ 12 –í?

–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (ùêª0): –°—Ä–µ–¥–Ω–µ–µ –Ω–∞–ø—Ä—è–∂–µ–Ω–∏–µ –±–∞—Ç–∞—Ä–µ–µ–∫ —Ä–∞–≤–Ω–æ 12–í.
–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H1): –°—Ä–µ–¥–Ω–µ–µ –Ω–∞–≤–ø—Ä—è–∂–µ–Ω–∏–µ –±–∞—Ç–∞—Ä–µ–µ–∫ –Ω–µ —Ä–∞–≤–Ω–æ 12–í.

```python
import numpy as np
from scipy import stats
```

```python
x = np.array([12.9, 11.6, 13.5, 13.9, 12.1, 11.9, 13])
```

```python
mu0 = 12
```

```python
df = len(x)-1
student = stats.t(df)
T = (df)**0.5*(x.mean()-mu0)/x.std()
P = student.cdf(T)
print('Statistic:', T, 'p-value', P)
```

–ü–æ —Ç–∞–±–ª–∏—Ü–µ t-—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–ª—è 6 —Å—Ç–µ–ø–µ–Ω–µ–π —Å–≤–æ–±–æ–¥—ã –∏ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º—Å–æ—Ç–∏ 0.05 –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–Ω–µ–Ω–∏–µ t —Ä–∞–≤–Ω–æ 2.45. –ü–æ–ª—É—á–µ–Ω–Ω–∞—è t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –º–µ–Ω—å—à–µ –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è, –ø–æ—ç—Ç–æ–º—É –º—ã –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ —Å—Ä–µ–¥–Ω–µ–µ –Ω–∞–ø—Ä—è–∂–µ–Ω–∏–µ —Ä–∞–≤–Ω–æ 12–í. –ö —Ç–æ–º—É –∂–µ –∑–Ω–∞—á–µ–Ω–∏–µ p-value –±–æ–ª—å—à–µ 0.05, —á—Ç–æ –Ω–µ –ø–æ–∑–≤–æ–ª—è–µ—Ç –Ω–∞–º –æ—Ç–≤—Ä–µ–≥–∞—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É. 

<b>–ó–∞–¥–∞–Ω–∏–µ 2.</b> –ï—Å—Ç—å –¥–≤–∞ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –º–µ—Ç–æ–¥–∞ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø—Ä–æ—Ü–µ–Ω—Ç–Ω–æ–≥–æ —Å–æ–¥–µ—Ä–∂–∞–Ω–∏—è –∂–∏—Ä–∞ –≤ –º—è—Å–µ. –î–æ–ª—è –∂–∏—Ä–∞ –¥–ª—è 10 –æ–±—Ä–∞–∑—Ü–æ–≤ –º—è—Å–∞ –±—ã–ª–∞ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞ –∫–∞–∂–¥—ã–º –∏–∑ –¥–≤—É—Ö –º–µ—Ç–æ–¥–æ–≤. –ü–æ–∑–≤–æ–ª—è—é—Ç –ª–∏ –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥ –æ —Ç–æ–º, —á—Ç–æ –æ–¥–∏–Ω –º–µ—Ç–æ–¥ –¥–∞—ë—Ç –≤ —Å—Ä–µ–¥–Ω–µ–º –±–æ–ª–µ–µ –≤—ã—Å–æ–∫–∏–µ –ø–æ–∫–∞–∑–∞–Ω–∏—è, —á–µ–º –¥—Ä—É–≥–æ–π?

–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H0): –°—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –¥–≤—É—Ö –º–µ—Ç–æ–¥–æ–≤ —Ä–∞–≤–Ω—ã (—Ä–∞–∑–Ω–∏—Ü–∞ —Ä–∞–≤–Ω–∞ 0).
–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H1): –°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ú–µ—Ç–æ–¥–∞ I –±–æ–ª—å—à–µ, —á–µ–º –ú–µ—Ç–æ–¥–∞ II.

```python
x = np.array([23.1, 23.2, 26.5, 26.6, 27.1, 48.3, 40.5, 25, 38.4, 23.5])
y = np.array([22.7, 23.6, 27.1, 27.4, 27.4, 46.8, 40.4, 24.9, 38.1, 23.8])

z = x-y
```

```python
df = len(x)-1
student = stats.t(df)
T = (df)**0.5*z.mean()/z.std()
P = student.cdf(T)
print('Statistic:', T, 'p-value', P)
```

–ü–æ—Å–∫–æ–ª—å–∫—É p-value –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –±–æ–ª—å—à–µ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 0.05, –º—ã –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É.
–ù–µ—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã—Ö —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–∏—Ö –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è—Ö –¥–≤—É—Ö –º–µ—Ç–æ–¥–æ–≤ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø—Ä–æ—Ü–µ–Ω—Ç–Ω–æ–≥–æ —Å–æ–¥–µ—Ä–∂–∞–Ω–∏—è –∂–∏—Ä–∞ –≤ –º—è—Å–µ.

<b>–ó–∞–¥–∞–Ω–∏–µ 3.</b> –î–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –≥–∏–ø–æ—Ç–µ–∑—ã –æ —Ç–æ–º, —á—Ç–æ –∂–∏—Ç–µ–ª–∏ –≥–æ—Ä–æ–¥–æ–≤ –∏–º–µ—é—Ç –≤ —Å—Ä–µ–¥–Ω–µ–º –±–æ–ª–µ–µ –≤—ã—Å–æ–∫–æ–µ —Å–∏—Å—Ç–æ–ª–∏—á–µ—Å–∫–æ–µ –∞—Ä—Ç–µ—Ä–∏–∞–ª—å–Ω–æ–µ –¥–∞–≤–ª–µ–Ω–∏–µ, —á–µ–º –∂–∏—Ç–µ–ª–∏ —Å–µ–ª—å—Å–∫–æ–π –º–µ—Å—Ç–Ω–æ—Å—Ç–∏, –±—ã–ª–æ –ø—Ä–æ–≤–µ–¥–µ–Ω–æ –æ–±—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ 10 –≥–æ—Ä–æ–∂–∞–Ω –∏ 9 –∂–∏—Ç–µ–ª–µ–π —Å–µ–ª–∞ –ø—Ä–∏–º–µ—Ä–Ω–æ –æ–¥–Ω–æ–≥–æ –≤–æ–∑—Ä–∞—Å—Ç–∞. –ü–æ–ª—É—á–µ–Ω—ã —Å–ª–µ–¥—É—é—â–∏–µ –¥–∞–Ω–Ω—ã–µ –æ–± —É—Ä–æ–≤–Ω–µ –¥–∞–≤–ª–µ–Ω–∏—è –¥–ª—è –≥–æ—Ä–æ–∂–∞–Ω: 132, 111, 119, 138, 200, 131, 138, 170, 159, 140; –¥–ª—è —Å–µ–ª—è–Ω: 115, 190, 127, 155, 148, 121, 116, 121, 197. –ú–æ–∂–Ω–æ –ª–∏ —Å—á–∏—Ç–∞—Ç—å –≤–µ—Ä–Ω–æ–π –≤—ã–¥–≤–∏–Ω—É—Ç—É—é –≥–∏–ø–æ—Ç–µ–∑—É –Ω–∞ —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 0,05?

–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H0): –°—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –∞—Ä—Ç–µ—Ä–∏–∞–ª—å–Ω–æ–≥–æ –¥–∞–≤–ª–µ–Ω–∏–µ —É –≥–æ—Ä–æ–∂–∞–Ω –∏ —É —Å–µ–ª—å—Å–∫–∏—Ö –∂–∏—Ç–µ–ª–µ–π —Ä–∞–≤–Ω—ã (—Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É —Å—Ä–µ–¥–Ω–∏–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ —Ä–∞–≤–Ω–∞ 0).
–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H1): –°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∞—Ä—Ç–µ—Ä–∏–∞–ª—å–Ω–æ–≥–æ –¥–∞–≤–ª–µ–Ω–∏–µ —É –≥–æ—Ä–æ–∂–∞–Ω –±–æ–ª—å—à–µ, —á–µ–º —É —Å–µ–ª—å—Å–∫–∏—Ö –∂–∏—Ç–µ–ª–µ–π.

```python
x = np.array([132, 111, 119, 138, 200, 131, 138, 170, 159, 140])
y = np.array([115, 190, 127, 155, 148, 121, 116, 121, 197])
```

```python
dfx = len(x)-1
dfy = len(y)-1
student = stats.t(dfx+dfy-2)
T = (x.mean()-y.mean())/(x.std()**2/dfx + y.std()**2/dfy)**0.5
P = student.cdf(T)
print('Statistic:', T, 'p-value', P)
```

–ó–Ω–∞—á–µ–Ω–∏–µ p_value –±–æ–ª—å—à–µ 0.05, —Ç–æ —É –Ω–∞—Å –Ω–µ—Ç –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –æ—Å–Ω–æ–≤–∞–Ω–∏–π –¥–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É. –¢–æ –µ—Å—Ç—å, —Ä–∞–∑–Ω–∏—Ü–∞ –≤ –¥–∞–≤–ª–µ–Ω–∏–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –Ω–µ–∑–Ω–∞—á–∏–º–∞.

<b>–ó–∞–¥–∞–Ω–∏–µ 4.</b> –ò–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ 52% –Ω–æ–≤–æ—Ä–æ–∂–¥—ë–Ω–Ω—ã—Ö –¥–µ—Ç–µ–π ‚Äî –º–∞–ª—å—á–∏–∫–∏, –∞ 48% ‚Äî –¥–µ–≤–æ—á–∫–∏. –í —Å–ª—É—á–∞–π–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –∏–∑ 5 000 —á–µ–ª–æ–≤–µ–∫ –≤ –≤–æ–∑—Ä–∞—Å—Ç–µ –æ—Ç 30 –¥–æ 60 –ª–µ—Ç –º—É–∂—á–∏–Ω –æ–∫–∞–∑–∞–ª–æ—Å—å 2 500. –ú–æ–∂–Ω–æ –ª–∏ –Ω–∞ —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 0,05 —Å—á–∏—Ç–∞—Ç—å, –æ—Å–Ω–æ–≤—ã–≤–∞—è—Å—å –Ω–∞ —ç—Ç–∏—Ö –¥–∞–Ω–Ω—ã—Ö, —á—Ç–æ —Å–º–µ—Ä—Ç–Ω–æ—Å—Ç—å —Å—Ä–µ–¥–∏ –º—É–∂—á–∏–Ω –∏ –∂–µ–Ω—â–∏–Ω –æ–¥–∏–Ω–∞–∫–æ–≤–∞?

–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H0): –°–º–µ—Ä—Ç–Ω–æ—Å—Ç—å —Å—Ä–µ–¥–∏ –º—É–∂—á–∏–Ω –∏ –∂–µ–Ω—â–∏–Ω –æ–¥–∏–Ω–∞–∫–æ–≤–∞.
–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (H1): –°–º–µ—Ä—Ç–Ω–æ—Å—Ç—å —Å—Ä–µ–¥–∏ –º—É–∂—á–∏–Ω –∏ –∂–µ–Ω—â–∏–Ω —Ä–∞–∑–ª–∏—á–Ω–∞.

```python
from scipy.stats import chisquare

# –ù–∞–±–ª—é–¥–∞–µ–º—ã–µ —á–∞—Å—Ç–æ—Ç—ã
observed = np.array([2500, 2500])
# –û–∂–∏–¥–∞–µ–º—ã–µ —á–∞—Å—Ç–æ—Ç—ã
expected = np.array([0.52 * 5000, 0.48 * 5000])

# –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ —Ç–µ—Å—Ç–∞ —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç
T, P = chisquare(observed, f_exp=expected)
print('Statistic:', T, 'p-value', P)
```

–ó–Ω–∞—á–µ–Ω–∏–µ p_value –º–µ–Ω—å—à–µ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 0.05, –ø–æ—ç—Ç–æ–º—É –º—ã –æ—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É.
–ù–∞ —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 0.05 –º—ã –º–æ–∂–µ–º —Å—á–∏—Ç–∞—Ç—å, —á—Ç–æ —Å–º–µ—Ä—Ç–Ω–æ—Å—Ç—å —Å—Ä–µ–¥–∏ –º—É–∂—á–∏–Ω –∏ –∂–µ–Ω—â–∏–Ω —Ä–∞–∑–ª–∏—á–∞–µ—Ç—Å—è.

<b>–ó–∞–¥–∞–Ω–∏–µ 5.</b> –í –ø–µ—Ä–≤–æ–º —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–µ –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –¥–≤—É—Ö —Å—Ä–µ–¥–Ω–∏—Ö X1=17, X2=16 –ø—Ä–∏–º–µ–Ω—è–ª—Å—è t-–∫—Ä–∏—Ç–µ—Ä–∏–π –°—Ç—å—é–¥–µ–Ω—Ç–∞, –∏ —ç—Ç–∏ —Ä–∞–∑–ª–∏—á–∏—è –æ–∫–∞–∑–∞–ª–∏—Å—å –∑–Ω–∞—á–∏–º—ã (p = 0, 001).

–í–æ –≤—Ç–æ—Ä–æ–º –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–∏, —Ç–∞–∫–∂–µ –ø—Ä–∏ –ø–æ–º–æ—â–∏ t - –∫—Ä–∏—Ç–µ—Ä–∏—è, —Å—Ä–∞–≤–Ω–∏–≤–∞–ª–∏—Å—å –¥–≤–∞ —Å—Ä–µ–¥–Ω–∏—Ö X1=17, X2=36, –∏ —ç—Ç–∏ —Ä–∞–∑–ª–∏—á–∏—è –Ω–µ –∑–Ω–∞—á–∏–º—ã –ø—Ä–∏ p = 0,8. –í —á–µ–º –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—Ä–∏—á–∏–Ω–∞ —Ç–∞–∫–∏—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤, –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏, —á—Ç–æ –Ω—É–ª–µ–≤–∞—è –∏ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑—ã –≤ –æ–±–æ–∏—Ö —Å–ª—É—á–∞—è—Ö –æ–¥–∏–Ω–∞–∫–æ–≤—ã?

* t-–∫—Ä–∏—Ç–µ—Ä–∏–π –°—Ç—å—é–¥–µ–Ω—Ç–∞ —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª–µ–Ω –∫ —Ä–∞–∑–º–µ—Ä—É –≤—ã–±–æ—Ä–æ–∫. –ë–æ–ª—å—à–∏–µ –≤—ã–±–æ—Ä–∫–∏ –º–æ–≥—É—Ç –¥–∞–≤–∞—Ç—å –±–æ–ª–µ–µ —Ç–æ—á–Ω—ã–µ –æ—Ü–µ–Ω–∫–∏ –∏, —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, –±–æ–ª–µ–µ –∑–Ω–∞—á–∏–º—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã.

* –ï—Å–ª–∏ –¥–∏—Å–ø–µ—Ä—Å–∏—è –≤ –æ–¥–Ω–æ–π –∏–∑ –≤—ã–±–æ—Ä–æ–∫ –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –±–æ–ª—å—à–µ, —ç—Ç–æ –º–æ–∂–µ—Ç –ø—Ä–∏–≤–µ—Å—Ç–∏ –∫ —Ç–æ–º—É, —á—Ç–æ —Ä–∞–∑–ª–∏—á–∏—è –º–µ–∂–¥—É —Å—Ä–µ–¥–Ω–∏–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –±—É–¥—É—Ç —Å—á–∏—Ç–∞—Ç—å—Å—è –Ω–µ–∑–Ω–∞—á–∏–º—ã–º–∏.
